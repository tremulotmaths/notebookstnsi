# SQLite et Python

## 1. Exemple avec définition des données

```python
# -*- coding:utf8 -*-

# Exemples SQLite et Python

import sqlite3 # importation du module

# création et connexion à la base

conn = sqlite3.connect('bddmeteo.db')

cur = conn.cursor() # création d'un curseur

# création de la première table (table Météo)

cur.execute("""
CREATE TABLE IF NOT EXISTS Météo(
Id INTEGER PRIMARY KEY AUTOINCREMENT UNIQUE,
Lieu TEXT, Année INTEGER, Mois TEXT,
Tmin INTEGER, Tmax INTEGER, Précipitations INTEGER)
"""
    )

conn.commit() # pour transmettre à la base de données

# définition des données (type tuple)

data = [(1, 'Antibes', 2017, 'Janvier', 7, 11, 19),
        (2, 'Antibes', 2017, 'Février', 10, 13, 18),
        (3, 'Antibes', 2017, 'Mars', 11, 15, 28),
        (4, 'Antibes', 2017, 'Avril', 13, 17, 18),
        (5, 'Antibes', 2017, 'Mai', 16, 20, 20),
        (6, 'Antibes', 2017, 'Juin', 21, 25, 8),
        (7, 'Antibes', 2017, 'Juillet', 22, 26, 2),
        (8, 'Antibes', 2017, 'Août', 23, 27, 3),
        (9, 'Antibes', 2017, 'Septembre', 18, 22, 16),
        (10, 'Antibes', 2017, 'Octobre', 17, 20, 1),
        (11, 'Antibes', 2017, 'Novembre', 12, 15, 15),
        (12, 'Antibes', 2017, 'Décembre', 8, 12, 38)]

# insertion des données

for d in data:
    cur.execute("""
    INSERT INTO Météo (Id, Lieu, Année, Mois, Tmin, Tmax, Précipitations)
    VALUES(?, ?, ?, ?, ?, ?, ?)
""",d)

conn.commit

# test d'une requête

R = "SELECT * FROM Météo;"
cur.execute(R)
for t in cur:
    print(t)

cur.close()
conn.close() # déconnexion
```

## 2. Exemple avec les données dans un fichier csv

```python
# -*- coding:utf8 -*-

# Exemples SQLite et Python

import sqlite3 # importation du module

# création et connexion à la base

conn = sqlite3.connect('bddmeteo2.db')

cur = conn.cursor() # création d'un curseur

# création de la première table (table Météo)

cur.execute("""
CREATE TABLE IF NOT EXISTS Météo(
Id INTEGER PRIMARY KEY AUTOINCREMENT UNIQUE,
Lieu TEXT, Année INTEGER, Mois TEXT,
Tmin INTEGER, Tmax INTEGER, Précipitations INTEGER)
"""
    )

conn.commit() # pour transmettre à la base de données

# utilisation d'un fichier csv pour les données

fic = open('data.csv', 'r', encoding='utf8')
fic.readline() # lecture et élimination de la première ligne inutile

donnees = []
for ligne in fic:
    donnees.append(ligne.rstrip().split(','))

# conversion string en int
for liste in donnees:
    liste[0] = int(liste[0])
    liste[2] = int(liste[2])
    liste[4] = int(liste[4])
    liste[5] = int(liste[5])
    liste[6] = int(liste[6])


# insertion des données

for d in donnees:
    cur.execute("""
    INSERT INTO Météo (Id, Lieu, Année, Mois, Tmin, Tmax, Précipitations)
    VALUES(?, ?, ?, ?, ?, ?, ?)
""",d)

conn.commit

# test d'une requête

R = "SELECT * FROM Météo;"
cur.execute(R)
for t in cur:
    print(t)

cur.close()
conn.close() # déconnexion
```
